```python
# -------------------------------
# STEP 3: DATA CLEANING
# -------------------------------

# Remove duplicate rows
df = df.drop_duplicates()
print("Data Cleaning Done ✅")
```

```python
# -------------------------------
# STEP 4: FIX DATA TYPES
# -------------------------------

# Convert Age to integer
df['Age'] = df['Age'].astype(int)

# Convert Monthly Income to float
df['Monthly_Income'] = df['Monthly_Income'].astype(int)
```

```python
# -------------------------------
# STEP 5: FEATURE ENGINEERING
# -------------------------------

# 1. Create Income Category column
# Low < 5000, Medium 5000–15000, High > 15000
def income_category(income):
    if income < 5000:
        return "Low"
    elif income <= 15000:
        return "Medium"
    else:
        return "High"
```

```python
# Apply function on Monthly_Income column
df['Income_Category'] = df['Monthly_Income'].apply(income_category)


# 2. Create Experience Level column
# Based on Years_at_Company
def experience_level(years):
    if years < 3:
        return "Junior"
    elif years <= 7:
        return "Mid"
    else:
        return "Senior"
```

```python
# Apply function
df['Experience_Level'] = df['Years_at_Company'].apply(experience_level)


# 3. Create Promotion Status column
# If employee working >=5 years → No Promotion
def promotion_status(years):
    if years >= 5:
        return "No Promotion 5+ Years"
    else:
        return "Recently Promoted"

# Apply function
df['Promotion_Status'] = df['Years_at_Company'].apply(promotion_status)
```

```python
df.columns
```

```python
df['Risk_Flag'] = "Normal"

df.loc[
    (
        (df['Attrition'] == 'Yes') &
        (df['Work_Environment_Satisfaction'] <= 2)
    ) |
    (
        (df['Promotion_Status'] == "No Promotion 5+ Years") &
        (df['Work_Environment_Satisfaction'] <= 2) &
        (df['Job_Involvement'] <= 2)
    ),
    'Risk_Flag'
] = "High Risk"
```

```python
# -------------------------------
# STEP 6: DATA VALIDATION
# -------------------------------

# Check unique values of new columns (for correctness)
print("Unique Income Categories:", df['Income_Category'].unique())
print("Unique Experience Levels:", df['Experience_Level'].unique())
print("Unique Promotion Status:", df['Promotion_Status'].unique())
print("Unique Risk Flags:", df['Risk_Flag'].unique())
```
